```python
#String, value, isOdd(1 == odd)
def plus():
    x = random.randint(0,100)
    y = random.randint(0,100)
    z = x+y
    z2 = z%2
    return str(x)+" + "+str(y), z ,z2
```

```python
def minus():
    x = random.randint(0,100)
    y = random.randint(0,100)
    z = x-y
    z2 = z%2
    return str(x)+" - "+str(y), z ,z2
```

```python
def multiply():
    x = random.randint(0,100)
    y = random.randint(0,100)
    z = x*y
    z2 = z%2
    return str(x)+" * "+str(y), z ,z2
```

```python
def divide():
    x = random.randint(0,100)
    y = random.randint(1,100)
    z = x//y
    z2 = z%2
    return str(x)+" // "+str(y), z ,z2
```

```python
def modulo():
    x = random.randint(0,100)
    y = random.randint(1,100)
    z = x%y
    z2 = z%2
    return str(x)+" % "+str(y), z ,z2
```

```python
def genPlus(number):
    oddList = []
    evenList = []
    while (len(oddList)<number or len(evenList)<number):
        gen = plus()
        if gen[2]==1:
            if len(oddList) != number:
                oddList.append(gen)
        else:
            if len(evenList) != number:
                evenList.append(gen)
    return evenList, oddList
```

```python
def genMinus(number):
    oddList = []
    evenList = []
    while (len(oddList)<number or len(evenList)<number):
        gen = minus()
        if gen[2]==1:
            if len(oddList) != number:
                oddList.append(gen)
        else:
            if len(evenList) != number:
                evenList.append(gen)
    return evenList, oddList
```

```python
def genMultiply(number):
    oddList = []
    evenList = []
    while (len(oddList)<number or len(evenList)<number):
        gen = multiply()
        if gen[2]==1:
            if len(oddList) != number:
                oddList.append(gen)
        else:
            if len(evenList) != number:
                evenList.append(gen)
    return evenList, oddList
```

```python
def genModulo(number):
    oddList = []
    evenList = []
    while (len(oddList)<number or len(evenList)<number):
        gen = modulo()
        if gen[2]==1:
            if len(oddList) != number:
                oddList.append(gen)
        else:
            if len(evenList) != number:
                evenList.append(gen)
    return evenList, oddList
```

```python
def genDivide(number):
    oddList = []
    evenList = []
    while (len(oddList)<number or len(evenList)<number):
        gen = divide()
        if gen[2]==1:
            if len(oddList) != number:
                oddList.append(gen)
        else:
            if len(evenList) != number:
                evenList.append(gen)
    return evenList, oddList
```
